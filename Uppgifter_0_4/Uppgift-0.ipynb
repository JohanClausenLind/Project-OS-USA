```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Uppgift 0

# 1. Read the data from the csv file into a pandas DataFrame.
file_path_athletes = '../Data/athlete_events.csv'
data_athletes = pd.read_csv(file_path_athletes)
# 2. Read the data from the csv file into a pandas DataFrame
file_regions = '../Data/noc_regions.csv'
regions = pd.read_csv(file_regions)
```

```python
#A. Hur många länder som är med i datan?
#B. Vilka länder är med?

# Count the number of NOCs in the data
countries = data_athletes['NOC'].nunique()
unique_countries = data_athletes['NOC'].drop_duplicates().dropna()
# Print the number of NOCs
print(f"Amount of countries: {countries}")
# Print the NOCs
print("All country abbreviations:", ", ".join(data_athletes['NOC'].drop_duplicates().dropna()))
```

```python
#Uppgift C # Visa unika sporter i datasetet

# 1. Get the unique values of the "Sport" column
unique_sports = data_athletes['Sport'].unique()
# 2. Get the length of the unique_sports array
num_sports = len(unique_sports)
# 3. Print the number of unique sports
print("Antal sporter i datasetet:",num_sports)
print()
# 4. Print the unique sports and join them with ", "
print("Sporter i datasetet:", ", ".join(unique_sports))
```

```python
# Uppgift D: # Visa antal medaljer per land

# 1. Drop rows with missing values in the Medal column
athletes_cleaned = data_athletes.dropna(subset=['Medal'])
# 2. Count the number of unique medals
medal_count = athletes_cleaned['Medal'].nunique()
# 3. Print the number of unique medals
print(f"Number of unique medals: {medal_count}")
```

```python
# Uppgift E: ta reda på statistik för åldern: medelvärde, median, min, max, standardavvikelse,
age_mean = data_athletes['Age'].mean()     # Mean (Medelvärde)
age_median = data_athletes['Age'].median() # Median (Median)
age_min = data_athletes['Age'].min()       # Minimum (Min)
age_max = data_athletes['Age'].max()       # Maximum (Max)
age_std_dev = data_athletes['Age'].std()   # Standard Deviation (Standardavvikelse)

# Print the values
print(f"Mean (Medelvärde): {age_mean}")
print(f"Median (Median): {age_median}")
print(f"Minimum (Min): {age_min}")
print(f"Maximum (Max): {age_max}")
print(f"Standard Deviation (Standardavvikelse): {age_std_dev}")
```

```python
#uppgift F: # Visa åldersfördelning bland deltagarna
# 1. Set figure size
plt.figure(figsize=(12, 6))
# 2. Create histogram of the Age column and drop missing values
sns.histplot(data_athletes['Age'].dropna(), bins=30, kde=True, color='skyblue')
# 3. Set title and labels
plt.title('Fördelning av åldrar bland deltagarna')
plt.xlabel('Ålder')
plt.ylabel('Antal deltagare')
plt.show() 
```

```python
#Uppgift G: # visar antal deltagare enlight kön

sns.countplot(data=data_athletes, x = 'Sex', palette=['blue', 'red'])

plt.title('Blanding mellan kön')
plt.xlabel('kön')
plt.ylabel('Antal män/kvinnor')
```

```python
#Uppgift H: # visar de 10 länderna med mest medaljer
medal_data = data_athletes.dropna(subset=['Medal'])

medal_count = medal_data.groupby('NOC')['Medal'].count().sort_values(ascending=False).head(10)

medal_count = medal_count.reset_index()

plt.figure(figsize = (10, 6))

sns.scatterplot(data = medal_count, x = 'NOC', y = 'Medal', s = 100)
                
plt.title('Top 10 länder med flest medaljer')
plt.xlabel('Land')
plt.ylabel('Antal medaljer')

plt.show()
```
